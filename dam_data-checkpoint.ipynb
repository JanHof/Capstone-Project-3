```python
group_by_country = dams_df.groupby(by=['Country'])

country_df = round(group_by_country.mean(),0)

dam_size_df = country_df.sort_values(by='Res_capacityM3', ascending=False)

# We will assemble a DataFrame of only the relevant features to plot  
features_of_interest = pd.DataFrame({'PV_InstalledCapacity': dam_size_df['PV_InstalledCapacity'],
                                     'Res_capacityM3': dam_size_df['Res_capacityM3'],})

## Let us plot a few of the manufactures for visibility
features_of_interest = features_of_interest.iloc[:10]
features_of_interest.plot(kind='barh', figsize=(20, 10))

plt.xlabel('Res Capacity', fontsize=12)
plt.ylabel('Country', fontsize=12)
```

Bar Chart compares the reservour capacity to the installed capacity.

```python
def plot_hist(variable):
    plt.figure(figsize = (9,3))
    plt.hist(dams_df[variable], bins = 50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()

```

```python
numericVar = ["Res_capacityM3", "PV_SpaCapacity","Res_area_km2"]
for n in numericVar:
    plot_hist(n)
```

Histograms showing the ditribution of 'Res_capacityM3', 'PV_SpaCapacity' and 'Res_area_km2'.

```python
purpose_count = dams_df.groupby('Purpose')['Purpose'].count()
purpose_count = purpose_count.sort_values(ascending=False)
purpose_count = purpose_count.iloc[:5]
# Displaying totals.
print(purpose_count)

# 'autopct' displays the percentages on the chart
plt.figure(figsize=(6,6))
plt.title('Purpose of Dams')
plt.pie(purpose_count.values, labels=purpose_count.index,autopct='%1.1f%%', textprops={'fontsize':16})
plt.show()
```

Pie chart shows the the percentage that applies to the purpose that the cyrrent dams are built for.

```python
plt.hist(dams_df['PV_InstalledCapacity'], 20, density =1,facecolor="blue", alpha=0.7)
plt.title('Installed Capacity')
plt.show()
```

```python
old_data = dams_df[dams_df['YEAR'] <2000 ].index
dams_df = dams_df.drop(old_data)
test = dams_df['Res_area_km2'].groupby(dams_df['YEAR']).size()
plot_year = test.plot(title = 'Annual Area')
plot_year.set_xlabel('Years')
plot_year.set_ylabel('km2')
```

Line graph shows the new area made available annually.

```python
x = dams_df['PotentialPVSurface'].values
y = dams_df['Res_area_km2'].values

plt.scatter(x,y,color = 'b')
plt.title('Potential PV Surface in Relation to Dam Area')
plt.xlabel('PotentialPVSurface ')
plt.ylabel('Res_area_km2')
plt.show()
```

Scatter plot shows correlation between potential v surface and reservoir area
